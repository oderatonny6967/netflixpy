```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
netfix = pd.read_csv("/content/sample_data/netflix_titles.csv")
```

```python
new_netflix= netfix.dropna(axis=0, inplace=False, how='any', subset=None)
```

```python
new_netflix.info()
```

```python
new_netflix.head()
```

```python
import matplotlib.pyplot as plt
```

```python
sns.boxplot(
    x = "release_year",
    y = "type",
    showmeans=True,  # Include means as markers within the boxes (optional)
    data=new_netflix
)

# Customize the plot (optional)
plt.xlabel('Year Released')
plt.ylabel('Content Type')
plt.title('Distribution of Content Type by Year Released')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python
content_by_country = new_netflix.groupby('country').agg(
    movies=pd.NamedAgg(column='type', aggfunc=lambda x: x.eq('Movie').sum()),
    tv_shows=pd.NamedAgg(column='type', aggfunc=lambda x: x.eq('Tv Show').sum())
)

# Create a bar graph
content_by_country.plot(kind='bar', stacked=False)
plt.xlabel('Countries')
plt.ylabel('Content Count')
plt.title('Content Availability by Country')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend(title='Content Type')  # Add a legend
plt.tight_layout()  # Adjust spacing for better readability
plt.show()
```

```python
palette = ['skyblue', 'lightcoral']
```

```python
sns.violinplot(
     x = "release_year",
    y = "type",
    #showmeans=True,  # Include means as markers within the boxes (optional)
    data=new_netflix,
    hue="type",  # Use "type" column for color differentiation
    palette=palette

)
plt.xlabel('Year Released')
plt.ylabel('Content Type')
plt.title('Distribution of Content Type by Year Released')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

```python
# Create a horizontal bar graph
content_by_country.plot(kind='barh', stacked=False)
plt.xlabel('Content Count')
plt.ylabel('Countries')
plt.title('Content Availability by Country')
plt.legend(title='Content Type')  # Add a legend
plt.tight_layout()  # Adjust spacing for better readability
plt.show()
```

```python
# Group content by country and count occurrences of 'movie' and 'tv' (assuming 'movie' and 'tv' represent counts)
content_by_country = new_netflix.groupby('country').size().to_frame(name='count')

# Calculate the number of countries (assuming you want to plot against the number of countries in each group)
content_by_country['num_countries'] = content_by_country.groupby('count')['count'].transform('size')

# Prepare the data for plotting
countries = content_by_country.index.to_numpy()  # Extract country labels
counts = content_by_country['count'].to_numpy()  # Extract content counts
num_countries = content_by_country['num_countries'].to_numpy()  # Extract number of countries per count

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size for better readability
plt.scatter(num_countries, counts)
plt.xlabel('Number of Countries')
plt.ylabel('Content Count (Assuming "movie" and "tv" represent counts)')
plt.title('Content Count vs. Number of Countries')
plt.grid(True)  # Add gridlines for better readability
plt.tight_layout()  # Adjust spacing for better readability
plt.show()
```

```python
# Group content by each country and get the content type ('movie' or 'tv')
content_by_country = new_netflix.groupby('country')['type'].apply(pd.Series.nunique)

# Extract specific countries and content counts
countries = content_by_country.index.to_numpy()  # Country labels
content_counts = content_by_country.to_numpy()  # Content counts (assuming 'movie' and 'tv' represent counts)

# Create the bar plot (or adjust for a different plot type)
plt.figure(figsize=(12, 6))  # Adjust figure size for better readability
plt.bar(countries, content_counts)
plt.xlabel('Countries')
plt.ylabel('Content Count (Assuming "movie" and "tv" represent counts)')
plt.title('Content Count by Specific Countries')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability if needed
plt.tight_layout()  # Adjust spacing for better readability
plt.show()
```
